```python
from sklearn.datasets import load_wine#Veri setlerini yüklemek için kullanılır. Burada şarap veri setini alacağız.
wine = load_wine()
X = wine.data#Girdi özelliklerini tutar (örneğin, asidite, alkol oranı vb.).
Y = wine.target#Y: Sınıf etiketlerini tutar (şarap türleri).
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split#Veri setini eğitim ve test olarak ayırmak için kullanılır.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)#RAndom state 
#random state Bu, kodun her çalıştırıldığında aynı bölme sonucunu vermesini sağlar.
```

```python
from sklearn.tree import DecisionTreeClassifier #Karar ağaçları ile çalışmak için gerekli sınıfları içerir.
dtc=DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
Y_predicition=dtc.predict(X_test)
Y_predicition
```

```python
from sklearn import metrics #Modelin performansını değerlendirmek için metrikleri sağlar.
result=metrics.accuracy_score(Y_test,Y_predicition)#accuracy_score fonksiyonu, şu adımları takip eder:

#Karşılaştırma: Y_test ile Y_prediction arasındaki değerleri karşılaştırır.
#Doğru Tahminler: Her bir öğe için, gerçek sınıfın tahmin edilen sınıf ile eşleşip eşleşmediğini kontrol eder.
#Eğer eşleşiyorsa, bu doğru bir tahmin olarak sayılır.
#result=dogru tahmin/toplam tahmin
#ınıflandırma problemlerinde modelin doğru tahmin oranını ölçmek için kullanılır.accuracy score.

result
```

```python
from sklearn.datasets import load_diabetes
diabetes=load_diabetes()
X=diabetes.data
Y=diabetes.target
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
#Scikit-learn, makine öğrenimi algoritmalarını uygulamak için yaygın olarak kullanılan bir Python kütüphanesidir.
#ensemble(grup): Bu modül, birden fazla öğrenme algoritmasını bir araya getirerek daha iyi tahminler yapmak için kullanılır.
#RandomForestRegressor: Rastgele Orman (Random Forest) algoritmasını kullanan bir regresyon modelidir.
#Bu model, birden fazla karar ağaçlarından oluşur ve bu ağaçların tahminlerini birleştirerek daha sağlam ve doğru sonuçlar elde etmeye çalışır.
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,max_features="sqrt",random_state=42)
model.fit(X_train,Y_train)
#n_estimators=100: Modelin oluşturacağı karar ağaçlarının sayısını belirtir.
#Burada 100 ağaç oluşturulacak. Daha fazla ağaç genellikle daha iyi bir performans sağlar, ancak hesaplama süresini artırabilir.
#max_features="sqrt": Her bir ağaç için rastgele seçilecek özelliklerin sayısını belirler.
#"sqrt" seçeneği, her ağaç için özelliklerin karekökü kadar özelliğin seçileceği anlamına gelir. 
#Bu, modelin çeşitliliğini artırarak overfitting (aşırı uyum) riskini azaltır.
#random_state=4: Rastgelelik içeren işlemlerin tutarlılığı için bir başlangıç değeri belirler. 
#Aynı random_state değeri ile model yeniden çalıştırıldığında, aynı sonuçlar elde edilir. Bu, sonuçların tekrarlanabilir olmasını sağlar.
```

```python
Y_prediction=model.predict(X_test)
Y_prediction
```

```python
from sklearn import metrics

mse=metrics.mean_squared_error(Y_test,Y_prediction)
mse
#hata=gercek_tahmin 1/n*hata '2
#Düşük bir MSE değeri, modelinin iyi performans gösterdiğini gösterir.



```

```python
import numpy  as np
rmse=np.sqrt(mse)
rmse

# kullanmamızın nedeni, sqrt() fonksiyonunun NumPy kütüphanesinde tanımlı olmasıdır.
#Eğer sadece sqrt() yazarsak, Python bu fonksiyonu tanımayacaktır, çünkü onu içe aktarmadık. 
```

```python
import pandas as pd
data="https://raw.githubusercontent.com/azat-bit/NovaAkademi/refs/heads/main/NovaAkademi/musteri_verileri.csv"
framework=pd.read_csv(data)
framework
```

```python
from sklearn.preprocessing import LabelEncoder
gender=LabelEncoder()
category_of_product=LabelEncoder()

```

```python
framework["Cinsiyet"]=gender.fit_transform(framework["Cinsiyet"])
#scikit-learn kütüphanesinin bir parçası olan ve kategorik değişkenleri sayısal formatta temsil etmek için kullanılan bir araçtır. 
#Özellikle makine öğrenimi algoritmaları, kategorik verilerle doğrudan çalışamazlar, bu yüzden bu verilerin sayısal hale dönüştürülmesi gerekir. 
#LabelEncoder bu işlevi yerine getirir.
framework["Urun_Kategorisi"]=category_of_product.fit_transform(framework["Urun_Kategorisi"])
framework
```

```python

X=framework.drop("Memnuniyet_Puani",axis=1)#Sutunları düşürmek için kullanılır .memnu.planını sutunu cıkar geri kalanı yaz
Y=framework["Memnuniyet_Puani"]# axis=0 Satırları düşürmek için kullanılır.
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
r_state=RandomForestRegressor(random_state=42,n_estimators=100)
r_state
#Scikit-learn, makine öğrenimi algoritmalarını uygulamak için yaygın olarak kullanılan bir Python kütüphanesidir.
#ensemble(grup): Bu modül, birden fazla öğrenme algoritmasını bir araya getirerek daha iyi tahminler yapmak için kullanılır.
#RandomForestRegressor: Rastgele Orman (Random Forest) algoritmasını kullanan bir regresyon modelidir.
#Bu model, birden fazla karar ağaçlarından oluşur ve bu ağaçların tahminlerini birleştirerek daha sağlam ve doğru sonuçlar elde etmeye çalışır.
```

```python
r_state.fit(X_train,Y_train)
```

```python
Y_prediction=r_state.predict(X_test)
Y_prediction
```

```python

from sklearn.metrics import mean_squared_error,r2_score
import numpy as np

mse=metrics.mean_squared_error(Y_test,Y_prediction)

print(mse,"mean squared error")




r2_score=r2_score(Y_test,Y_prediction)
print(r2_score,"R-SQUARED")


# r kare 0: Model, bağımlı değişkenin varyansını açıklayamıyor. Yani, model tamamen başarısız.
#1: Model, bağımlı değişkenin varyansını mükemmel bir şekilde açıklıyor. Yani, model mükemmel
#Negative Value: The model performs worse than a simple mean prediction, suggesting a very poor fit.
```
